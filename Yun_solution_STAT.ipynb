```python
import numpy as np
import pandas as pd
import requests
```

```python
response=requests.get('https://live.markethound.in/api/history/dtelist?index=FINNIFTY&expiry=2024-03-05T00:00:00.000Z')
```

```python
df=pd.DataFrame(response.json()['result'])
df
```

```python
df_new=df[['date','close','high','low','open']]
df_new
```

```python
df_new['date'] = pd.to_datetime(df_new['date'])
df_new['date'] = df_new['date'].dt.strftime('%Y-%m-%d')
df_new
```

```python
df_new = df_new.sort_values(by='date')
df_new
```

```python
from sklearn.preprocessing import MinMaxScaler
import tensorflow
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
def create_dataset(dataset, time_step=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step-1):
		a = dataset[i:(i+time_step), 0]
		dataX.append(a)
		dataY.append(dataset[i + time_step, 0])
	return np.array(dataX), np.array(dataY)
```

```python
#creating model for predicting Close name as 'model_1'
df1=df_new['close']
scaler=MinMaxScaler(feature_range=(0,1))
df1=scaler.fit_transform(np.array(df1).reshape(-1,1))
X_train_1, y_train_1 = create_dataset(df1, 3)
X_train_1 =X_train_1.reshape(X_train_1.shape[0],X_train_1.shape[1] , 1)

#model
model_1=Sequential()
model_1.add(LSTM(5,return_sequences=True,input_shape=(3,1)))
model_1.add(LSTM(5))
model_1.add(Dense(1))
model_1.compile(loss='mean_squared_error',optimizer='adam')
model_1.fit(X_train_1,y_train_1,epochs=50,batch_size=1)
```

```python
#creating model for predicting Open name as 'model_2'
df2=df_new['open']
scaler=MinMaxScaler(feature_range=(0,1))
df2=scaler.fit_transform(np.array(df2).reshape(-1,1))
X_train_2, y_train_2 = create_dataset(df2, 3)
X_train_2 =X_train_2.reshape(X_train_2.shape[0],X_train_2.shape[1] , 1)

#model
model_2=Sequential()
model_2.add(LSTM(5,return_sequences=True,input_shape=(3,1)))
model_2.add(LSTM(5))
model_2.add(Dense(1))
model_2.compile(loss='mean_squared_error',optimizer='adam')
model_2.fit(X_train_2,y_train_2,epochs=50)
```

```python
#creating model for predicting High name as 'model_3'
df3=df_new['high']
scaler=MinMaxScaler(feature_range=(0,1))
df3=scaler.fit_transform(np.array(df3).reshape(-1,1))
X_train_3, y_train_3 = create_dataset(df3, 3)
X_train_3 =X_train_3.reshape(X_train_3.shape[0],X_train_3.shape[1] , 1)

#model
model_3=Sequential()
model_3.add(LSTM(5,return_sequences=True,input_shape=(3,1)))
model_3.add(LSTM(5))
model_3.add(Dense(1))
model_3.compile(loss='mean_squared_error',optimizer='adam')
model_3.fit(X_train_3,y_train_3,epochs=50,batch_size=1)
```

```python
#creating model for predicting Low name as 'model_4'
df4=df_new['low']
scaler=MinMaxScaler(feature_range=(0,1))
df4=scaler.fit_transform(np.array(df4).reshape(-1,1))
X_train_4, y_train_4 = create_dataset(df4, 3)
X_train_4 =X_train_4.reshape(X_train_4.shape[0],X_train_4.shape[1] , 1)

#model
model_4=Sequential()
model_4.add(LSTM(5,return_sequences=True,input_shape=(3,1)))
model_4.add(LSTM(5))
model_4.add(Dense(1))
model_4.compile(loss='mean_squared_error',optimizer='adam')
model_4.fit(X_train_4,y_train_4,epochs=50,batch_size=1)
```

```python
y_pred_1_=model_1.predict(X_train_1)
y_pred_2_=model_2.predict(X_train_2)
y_pred_3_=model_3.predict(X_train_3)
y_pred_4_=model_4.predict(X_train_4)

y_pred_1=np.array(scaler.inverse_transform(y_pred_1_))
y_pred_2=np.array(scaler.inverse_transform(y_pred_2_))
y_pred_3=np.array(scaler.inverse_transform(y_pred_3_))
y_pred_4=np.array(scaler.inverse_transform(y_pred_4_))

true_values_1 = np.array(y_train_1)
true_values_2 = np.array(y_train_2)
true_values_3 = np.array(y_train_3)
true_values_4 = np.array(y_train_4)

mae_1 = np.mean(np.abs(y_pred_1 - true_values_1))
mae_2 = np.mean(np.abs(y_pred_2 - true_values_2))
mae_3 = np.mean(np.abs(y_pred_3 - true_values_3))
mae_4 = np.mean(np.abs(y_pred_4 - true_values_4))
```

```python
print('mean squared error for model 1 is:',mae_1,'\nmean squared error for model 2 is:',mae_2,'\nmean squared error for model 3 is:',mae_3,'\nmean squared error for model 4 is:',mae_4)
```
